```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

ax = plt.gca()

#triangle
pts = np.array([[0,0],[7,24],[14,0]])
p = Polygon(pts, closed=True, fc='white', edgecolor='black')
ax.add_patch(p)

#line
x1, y1 = [7,7],[0,24]
plt.plot(x1,y1,color = "black",linestyle='dotted')

#Dimensions of the figures
plt.text(3,8,"h=24 cm")
plt.text(8,-2,"r=7 cm")
plt.text(11,13, "l=25 cm")

#ellipse
u=7 #x-position of the center
v=0 #y-position of the center
a=7 #semi major axis
b=4#semi minor axis
t = np.linspace(0, 2*np.pi, 100)
plt.plot( u+a*np.cos(t) , v+b*np.sin(t), color='black')

#plotting center points
plt.plot(7, 0, marker="o", markersize=2.5, markeredgecolor="black", markerfacecolor="black")

ax.set_xlim(-5,20)
ax.set_ylim(-10,30)

plt.show()
```
