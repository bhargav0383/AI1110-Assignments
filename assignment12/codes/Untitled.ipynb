```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.array([-3,-2,-1,0, 1, 2, 3])
y = np.array([-1,-1,1,-1, 1, -1, 1])

plt.axvline(x=0, c="black", label="x=0")
plt.axhline(y=0, c="black", label="y=0")

plt.step(x, y)
plt.xlabel(r'$\omega$')
plt.ylabel("B($\omega$)")
plt.show()
```
